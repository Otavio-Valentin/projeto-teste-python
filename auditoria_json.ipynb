```python
import pandas as pd
import json

# 1. Criando os dados de teste (Sandbox)
mock_data = {
    "items": [
        {"type": 1, "name": "Google", "login": {"username": "otavio@gmail.com", "password": "abc"}},
        {"type": 1, "name": "GitHub", "login": {"username": "dev_valentin", "password": "123"}},
        {"type": 2, "name": "Cartão Antigo", "card": {"brand": "Master", "number": "5555"}},
        {"type": 3, "name": "Nota de Estudo", "notes": "Focar em Data Science na Cruzeiro"}
    ]
}

# 2. Transformando em DataFrame (Flattening)
df = pd.json_normalize(mock_data['items'])

# 3. Exibindo para confirmar que funcionou
print("Sucesso! A tabela 'df' está carregada na memória.")
display(df.head())
```

```python
# Agora o Python vai encontrar o 'df' porque ele não foi deletado
apenas_logins = df[df['type'] == 1]

print("--- Lista de Logins Isolada ---")
display(apenas_logins[['name', 'login.username']])
```

```python
# Vamos buscar qualquer item onde o nome contenha a palavra "Git"
busca = "Google"
resultado = df[df['name'].str.contains(busca, case=False)]

print(f"--- Resultados para a busca: '{busca}' ---")
display(resultado[['name', 'login.username']])
```

```python
# Criando uma cópia limpa do DataFrame para não alterar o original
df_limpo = df.copy()

# Substituindo todos os NaN por "Não se aplica"
df_limpo = df_limpo.fillna("N/A")

print("--- Tabela com Dados Limpos (Sem NaN) ---")
display(df_limpo)
```

```python
# 1. Selecionando apenas colunas seguras para o relatório
colunas_publicas = ['name', 'type', 'login.username']
df_relatorio = df_limpo[colunas_publicas]

# 2. Salvando em um arquivo CSV (aberto por qualquer planilha)
df_relatorio.to_csv('relatorio_auditoria_limpo.csv', index=False, encoding='utf-8')

# 3. Se você tiver a biblioteca 'openpyxl' instalada, pode salvar em Excel também:
# df_relatorio.to_excel('relatorio_auditoria_limpo.xlsx', index=False)

print("--- SUCESSO ---")
print("O arquivo 'relatorio_auditoria_limpo.csv' foi criado na sua pasta.")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Garantindo que os dados existam (Recriando o df e a contagem)
mock_data = {
    "items": [
        {"type": 1, "name": "Google", "login": {"username": "otavio@gmail.com"}},
        {"type": 1, "name": "GitHub", "login": {"username": "dev_valentin"}},
        {"type": 2, "name": "Cartão Antigo", "card": {"brand": "Master"}},
        {"type": 3, "name": "Nota de Estudo", "notes": "Data Science na Cruzeiro"}
    ]
}
df = pd.json_normalize(mock_data['items'])
contagem = df['type'].value_counts()

# 2. Configurando o Gráfico de Pizza
labels = ['Logins', 'Cartões', 'Notas']
cores = ['#4CAF50', '#FF5722', '#2196F3'] # Verde, Laranja, Azul

plt.figure(figsize=(7, 7))
plt.pie(contagem, labels=labels, autopct='%1.1f%%', startangle=140, colors=cores, shadow=True)
plt.title('Proporção de Ativos por Categoria', fontsize=15)

# 3. Exibindo
plt.show()
```
